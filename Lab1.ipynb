```python
def function2(n):
    return math.sinh(sum(random.random() for _ in range(n)) / n)
```

```python
def function3(n):
  return sum(math.sinh(random.random()) for _ in range(n)) / n
```

```python
def function4(n):
  return sum(sum(random.random() for _ in range(n)) for _ in range(n))
```

```python
def function2_np(n):
  return np.sinh((1/n) * np.sum(np.random.rand(n)))
```

```python
def function3_np(n):
  return 1/n * np.sum(np.sinh(np.random.rand(n)))
```

```python
def function4_np(n):
  return (1/n**2) * np.sum(np.sum(np.random.rand(n,n)))
```
